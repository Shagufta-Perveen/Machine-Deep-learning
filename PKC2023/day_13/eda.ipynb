# Exploratary Data Analysis
 Exploratary Data Analysis(EDA) is a method of `analysis` and `summarize` the data.In python we use pandas library to preform EDA on a dataset.



```python
# import libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#import dataset
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#summarize data
df.describe()
```

```python
# find the null values
df.isnull().sum()
```

```python
df.isnull().sum()*100/len(df)
```

```python
# finding unique values
df.nunique()
```

count null value is explore data and deal with null values is data wrangling.

```python
# visualize the nan values in heatmap
sns.heatmap(df.isnull(), yticklabels=False,cbar=False,cmap='viridis')
```

```python
df.drop(columns='deck', inplace=True)
```

```python
sns.countplot(data=df,x='survived')
```

```python
sns.barplot(data=df,x='survived',y='age',hue='who')
# in barplot at x-axis catagorical value and y-axis numerical value
```

```python
sns.countplot(x='pclass',data=df)
# in count plot we use catagorical values
```

```python
# data disturbution
sns.histplot(data=df,x='age',bins=20)
```

```python
sns.histplot(data=df,x='age',bins=30)
```

```python
df['age'].hist(bins=30,color='darkred',alpha=0.75)
#alpha parameter controls the transparency level of the plot elements.
# histplot always makea numerical data which have more unique values
```

```python
sns.boxplot(data=df,y='age',x='survived')
# in boxplot we use one axis numerical value an other axis catagorical value
```

```python
sns.boxplot(data=df,y='age',x='survived',hue='who')
```

```python
sns.scatterplot(x='age',y='fare', data=df)
#both axis has numerical data
plt.ylim(0,250)
```

```python
sns.scatterplot(x='age',y='fare', data=df,hue='pclass') # data not normalization 
```

```python
sns.pairplot(df, hue='survived')
```

```python
# stacked barplot/chart
df.groupby(['pclass','who'])['fare'].mean().unstack().plot(kind='bar',stacked=True)
```

```python
sns.swarmplot(data=df,x='survived',y='fare')
```

```python
sns.swarmplot(data=df,x='survived',y='age',hue='who')
```

```python
sns.catplot(x='pclass',y='fare', hue='embarked',kind='point',data=df)
```

```python
 #relation of two numeric variable
sns.lmplot(data=df, x='age',y='fare')
```

```python

#df = df[df['fare'] < 300] # do yourself

# Create the lmplot using the filtered DataFrame
sns.lmplot(data=df[df['fare'] < 300], x='age', y='fare',hue='pclass')

```

```python
sns.lmplot(data=df, x='age',y='fare',hue='pclass')
plt.ylim(0,300)
```

```python
# how to add a column and make a age group
sns.histplot(data=df,x='age')
```

```python
df['age_groups']=pd.cut(df['age'],[0,10,18,30,50,80],
                        labels=['Kids','Child','Youth','Middle age','Old'])
```

```python
df.head()
```

```python
df.groupby('age_groups')['survived'].mean().plot(kind='bar')
```

```python
df.groupby('age_groups')['fare'].mean().plot(kind='bar')
```

```python
sns.pairplot(df, hue='survived',vars=['age','fare','pclass'])
```
